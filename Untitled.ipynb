```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
volume_resistivity = np.loadtxt('volume_resistivity_data.txt')
X = volume_resistivity[:,0].reshape(-1,1)
y = np.log(volume_resistivity[:,1]).reshape(-1,1)
reg = LinearRegression().fit(X,y)
m = reg.coef_[0][0]
n = reg.intercept_[0]
print('Coeficiente R^2: ' + str(reg.score(X,y)))
print('Pendiente: ' + str(m))
print('Intersección: '+ str(n))
```

```python
f, ax = plt.subplots(1,1, figsize=(7,5))
ax.plot(X,y,'o--', label='Datos')
ax.plot(X, m*X+n, '-.', label='Ajuste Lineal')
ax.set_xlabel('Tiempo [h]')
ax.set_ylabel(r'log-Resistividad volumétrica [$\Omega\cdot $cm]')
ax.set_title('Ajuste Lineal Degradación de Resistividad Volumétrica', fontweight='bold')
ax.legend()
ax.text(3000,31.7,'ln(y)={:.3}x + ln({:.6})'.format(m,n))
ax.grid()
f.savefig('resistividad_volumetrica_ajuste.png', format='png')
```

```python
dielectric_strength = np.loadtxt('dielectric_strength_data.txt')
X_ = dielectric_strength[:,0].reshape(-1,1)
y_ = dielectric_strength[:,1].reshape(-1,1)
reg_ = LinearRegression().fit(X_,y_)
m_ = reg_.coef_[0][0]
n_ = reg_.intercept_[0]
print('Coeficiente R^2: ' + str(reg_.score(X_,y_)))
print('Pendiente: ' + str(m_))
print('Intersección: '+ str(n_))
```

```python
f, ax = plt.subplots(1,1, figsize=(7,5))
ax.plot(X_,y_,'o--', label='Datos')
ax.plot(X_, m_*X_+n_, '-.', label='Ajuste Lineal')
ax.set_xlabel('Tiempo [h]')
ax.set_ylabel('Fuerza dieléctrica [kV/mm]')
ax.set_title('Ajuste Lineal Degradación de la Fuerza Dieléctrica', fontweight='bold')
ax.legend()
ax.text(3000,43.5,'y={:.3}x + {:.6}'.format(m_,n_))
ax.grid()
f.savefig('fuerza_dielectrica_ajuste.png', format='png')
```

```python
V = 0.38 #en kV (tensión nominal entre fases)
h = 4.4 # en mm (2 veces la aislación por conductor) 
x_falla = (V/h-n_)/m_ # horas desde t=0 para que alcance este umbral
print('El cable falla luego de : {:.6} horas. Aproximadamente {:.4} años.'.format(x_falla, x_falla/8760))
```
